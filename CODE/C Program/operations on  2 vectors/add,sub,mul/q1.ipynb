## Q1: Add/Subtract/element-wise multiply of two vectors hA & hB, put results in hC 

```python
%%writefile q1.c

#include <stdio.h>
#include <stdlib.h>
#include <time.h>

#define N 60
#define LOW 10
#define HIGH 99

// ── CPU Functions (replace CUDA kernels)

// Addition
void vecAdd(int *A, int *B, int *C, int n)
{
    for(int i = 0; i < n; i++)
        C[i] = A[i] + B[i];
}

// Subtraction
void vecSub(int *A, int *B, int *C, int n)
{
    for(int i = 0; i < n; i++)
        C[i] = A[i] - B[i];
}

// Multiplication
void vecMul(int *A, int *B, int *C, int n)
{
    for(int i = 0; i < n; i++)
        C[i] = A[i] * B[i];
}

// ── Main
int main()
{
    int *A, *B, *C;

    A = (int*) malloc(N * sizeof(int));
    B = (int*) malloc(N * sizeof(int));
    C = (int*) malloc(N * sizeof(int));

    srand(time(NULL));

    // Initialize
    for(int i = 0; i < N; i++)
    {
        A[i] = rand()%90 + 10;
        B[i] = rand()%90 + 10;
    }

    // Print A
    printf("A: ");
    for(int i = 0; i < N; i++) printf("%4d", A[i]);
    printf("\n");

    // Print B
    printf("B: ");
    for(int i = 0; i < N; i++) printf("%4d", B[i]);
    printf("\n");

    // Addition
    vecAdd(A, B, C, N);
    printf("\nC = A + B:\n");
    for(int i = 0; i < N; i++) printf("%4d", C[i]);
    printf("\n");

    // Subtraction
    vecSub(A, B, C, N);
    printf("\nC = A - B:\n");
    for(int i = 0; i < N; i++) printf("%4d", C[i]);
    printf("\n");

    // Multiplication
    vecMul(A, B, C, N);
    printf("\nC = A * B:\n");
    for(int i = 0; i < N; i++) printf("%4d", C[i]);
    printf("\n");

    free(A);
    free(B);
    free(C);

    return 0;
}
```
