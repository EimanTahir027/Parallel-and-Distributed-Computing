```python
%%writefile q3.c

#include <stdio.h>
#include <stdlib.h>
#include <time.h>

#define N    60
#define LOW  10
#define HIGH 99

long long dotProduct(int* hA, int* hB, int n);
void printVec(const char* label, int* v, int n);

int main()
{
    int *hA, *hB;
    long long hProd;

    hA = (int*) malloc(N * sizeof(int));
    hB = (int*) malloc(N * sizeof(int));

    srand(time(NULL));             /* seed once, OUTSIDE loop */
    for(int i=0; i<N; i++)        /* ; not , */
    {
        hA[i] = rand()%90+10;     /* rand() not srand(), no * , hA not h_A */
        hB[i] = rand()%90+10;     /* rand() not srand(), no * , hB not h_B */
    }

    printVec("hA", hA, N);
    printVec("hB", hB, N);

    hProd = dotProduct(hA, hB, N);
    printf("\nDot Product (hProd) = %lld\n", hProd);

    free(hA);
    free(hB);
    return 0;
}

long long dotProduct(int* hA, int* hB, int n)
{
    long long hProd = 0;
    for(int i=0; i<n; i++)
        hProd += (long long)hA[i] * hB[i];
    return hProd;
}

void printVec(const char* label, int* v, int n)
{
    printf("\n%s:\n", label);
    for(int i=0; i<n; i++)
        printf("%4d", v[i]);
    printf("\n");
}
```
