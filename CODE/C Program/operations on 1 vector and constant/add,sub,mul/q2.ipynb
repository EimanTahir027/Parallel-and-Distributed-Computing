```python
%%writefile q2.c

#include <stdio.h>
#include <stdlib.h>
#include <time.h>

#define N    60
#define LOW  10
#define HIGH 99
#define K    12

void numAdd(int* hA, int* hC, int k, int n);
void numSub(int* hA, int* hC, int k, int n);
void numMul(int* hA, int* hC, int k, int n);
void printVec(const char* label, int* v, int n);

int main()
{
    int *hA, *hC;

    hA = (int*) malloc(N * sizeof(int));
    hC = (int*) malloc(N * sizeof(int));

    srand(time(NULL));             /* seed once, OUTSIDE loop */
    for(int i=0; i<N; i++)        /* semicolons ; not commas , */
    {
        hA[i] = rand()%90+10;     /* rand() not srand(), no * needed */
    }

    printVec("hA", hA, N);

    numAdd(hA, hC, K, N);
    printVec("hC = hA + 12", hC, N);

    numSub(hA, hC, K, N);
    printVec("hC = hA - 12", hC, N);

    numMul(hA, hC, K, N);
    printVec("hC = hA * 12", hC, N);

    free(hA);
    free(hC);
    return 0;
}

void numAdd(int* hA, int* hC, int k, int n)
{
    for(int i=0; i<n; i++)
        hC[i] = hA[i] + k;
}

void numSub(int* hA, int* hC, int k, int n)
{
    for(int i=0; i<n; i++)
        hC[i] = hA[i] - k;
}

void numMul(int* hA, int* hC, int k, int n)
{
    for(int i=0; i<n; i++)
        hC[i] = hA[i] * k;
}

void printVec(const char* label, int* v, int n)
{
    printf("\n%s:\n", label);
    for(int i=0; i<n; i++)
        printf("%4d", v[i]);
    printf("\n");
}
```
